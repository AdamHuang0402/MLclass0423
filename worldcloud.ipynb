```
import pandas as pd 
fname = pd.read_csv("drive/My Drive/AI_class_project/data_v3/apple_01_tag.csv")
fname
```

```
import os
import jieba
from urllib.request import urlretrieve

# 使用大辭典來做分詞
BIG_DICT_PATH = "dict.txt.big"
if not os.path.exists(BIG_DICT_PATH):
    print(BIG_DICT_PATH, "不存在, 幫你下載")
    url = "https://github.com/fxsjy/jieba/raw/master/extra_dict/dict.txt.big"
    urlretrieve(url, BIG_DICT_PATH)
else:
    print(BIG_DICT_PATH, "已經存在")
jieba.set_dictionary(BIG_DICT_PATH)

def newscut(s):
    s = str(s)
    s = " ".join(jieba.cut(s))
    s = s.replace("\r", "").replace("\n", "")
    return s
# pandas表格: DataFrame 一欄或者一列: Series
# Series.apply(功能名字)
# x_train = fname["內文"].apply(newscut)
x_test = fname["使用者評論"].apply(newscut)
x_test
```

```
x_test[1]
```

```
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS

# x_test[1] 是輸入的序列，需要經過分詞處理。
# 字體 SourceHanSansTW-Regular.otf 要上傳colab sever內
font = r'/content/SourceHanSansTW-Regular.otf'
if(font):
  print("true")
# Generate a word cloud image
wordcloud = WordCloud(background_color="white",font_path= font).generate(x_test[1])

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```

```
wordcloud.to_file('word_cloud.png')
```
