```
import pandas as pd
neg = pd.read_table("drive/My Drive/AI_class_project/neg_01.txt")
neg
```

```
pos = pd.read_table("drive/My Drive/AI_class_project/pos_01.txt")
pos
```

```
import numpy as np

neg_index = pd.DataFrame({"target":np.zeros(len(neg["content"]),int), "content":neg["content"]})
neg_index
```

```
pos_index = pd.DataFrame({"target":np.ones(len(pos["content"]),int), "content":pos["content"]})
pos_index
```

```
test_neg = pd.DataFrame({"target":neg_index["target"][:10000], "content":neg_index["content"][:10000]})
train_neg = pd.DataFrame({"target":neg_index["target"][10000:100000], "content":neg_index["content"][10000:100000]})
train_neg
```

```
test_pos = pd.DataFrame({"target":pos_index["target"][:10000], "content":pos_index["content"][:10000]})
train_pos = pd.DataFrame({"target":pos_index["target"][10000:100000], "content":pos_index["content"][10000:100000]})
train_pos
```

```
test_data = pd.concat([test_neg, test_pos], axis=0, ignore_index=True)
test_data
```

```
train_data = pd.concat([train_neg, train_pos], axis=0, ignore_index=True)
train_data
```

```
train_df = pd.DataFrame({"target":train_data["target"], "content":train_data["content"]})
train_df
```

```
test_df = pd.DataFrame({"target":test_data["target"], "content":test_data["content"]})
test_df
```

```
import jieba 
def poemcut(s):
    s = " ".join(jieba.cut(s))
    s = s.replace("\r", " ").replace("\n", " ")
    # s = s.replace(",", " ").replace("。", " ").replace("!", " ").replace("?", " ").replace("...", " ")
    # s = s.replace("(", " ").replace(")", " ").replace("[", " ").replace("]", " ").replace('"', " ").replace(':', " ").replace(':', " ")
    return s
# pandas表格: DataFrame 一欄或者一列: Series
# Series.apply(功能名字)
train_df["content"] = train_df["content"].apply(poemcut)
test_df["content"] = test_df["content"].apply(poemcut)
test_df["content"]
```

```
# # 預處理1. 先把文字化成數字
from tensorflow.keras.preprocessing.text import Tokenizer
# 出現太少的詞, 你可以選擇不看, 只留出現次數最高的2000(num_words)
tok = Tokenizer(num_words=2000,
                filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',
                lower=True,
                split=" ") #2000
tok.fit_on_texts(train_df["content"])
```

```
tok.word_index
```

```
tok.word_index['忙']
```

```
somestr = ['我 會 跟 你 說','的 是 了 我']
tok.texts_to_sequences(somestr)
```

```
train_df["content"][0]
```

```
x_train_seq = tok.texts_to_sequences(train_df["content"])
x_test_seq = tok.texts_to_sequences(test_df["content"])
pd.DataFrame(x_train_seq)
```

```
from tensorflow.keras.preprocessing.sequence import pad_sequences
x_train_pad = pad_sequences(x_train_seq, maxlen=256)
x_test_pad = pad_sequences(x_test_seq, maxlen=256)
pd.DataFrame(x_train_pad)
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Flatten, Dropout, Dense

layers = [
    # 2001 * 64 = 128064
    Embedding(2001, 64, mask_zero=True, input_length=256),
    Flatten(),
    Dense(256, activation="relu"),
    Dropout(0.25),
    Dense(2, activation="softmax")
]
model = Sequential(layers)
model.summary()
```

```
from tensorflow.keras.losses import SparseCategoricalCrossentropy
model.compile(loss=SparseCategoricalCrossentropy(),
              optimizer="adam",
              metrics=["accuracy"])
```

```
import numpy as np
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
callbacks = [
    EarlyStopping(patience=3, restore_best_weights=True),
]
model.fit(x_train_pad,
          np.array(train_df["target"]),
          batch_size=200,
          epochs=100,
          validation_split=0.1,
          verbose=2,
          callbacks=callbacks)
```

```
model.evaluate(x_test_pad, np.array(test_df["target"]))
```

```
layers = [
    # 2001 * 64 = 128064
    Embedding(2001, 64, mask_zero=True),
]
embedding = Sequential(layers)
w = model.layers[0].get_weights()
embedding.set_weights(w)
embedding.summary()
```
